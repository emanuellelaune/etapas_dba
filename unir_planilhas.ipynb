```python
import pandas as pd
```

```python

planilha1 = pd.read_excel('.xlsx')
planilha2 = pd.read_excel('.xlsx')
planilha3 = pd.read_excel('.xlsx')

# Passo 1: União de duas planilhas diferentes
print("Unindo planilhas")
planilha_completa = pd.concat([planilha1, planilha2], ignore_index=True)
```

```python
print(f"Total de dados atualmente: {len(planilha_completa)}")
print(f"Total de clientes únicos nos reports: {planilha_completa['ID_CLIENTE'].nunique()}")

# Salvar a planilha de reports unida (opcional)
planilha_completa.to_excel('.xlsx', index=False)
print("Planilha de reports unida salva: .xlsx")

# Passo 2:Juntar com uma terceira planilha, contanto que obedeça a uma condição
print("\nUnindo com outra planilha...")

# Inner join: mantém apenas os dados da coluna unida se obedeceram a condição
planilha_final = pd.merge(
    planilha_completa,
    planilha3,
    on='#CONDIÇÃO',
    how='inner'  # apenas valores que aparecem nas duas planilhas
)
```

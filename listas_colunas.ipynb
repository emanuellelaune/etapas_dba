Esse código serve para poder ler todas as informações existentes dentro de uma coluna

```python
import pandas as pd
```

```python
df = pd.read_excel(".xlsx")
#name_sheet = "Página 1" caso o arquivo tenha mais de uma planilha
print("COLUNAS")
print(df.columns)
#Caso queira ver as categorias unicas
categorias_unicas = df[''].unique()
print("\nTodas as categorias únicas em NOME_COLUNA:")
for cat in categorias_unicas:
    print(cat)
```

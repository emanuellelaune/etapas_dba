```python
import pandas as pd



df23 = pd.read_excel(r"C:\Users\ResTIC16\Documents\registro_atendimento_2023.xlsx")
df24 = pd.read_excel(r"C:\Users\ResTIC16\Documents\registros_2024.xlsx")
df25 = pd.read_excel(r"C:\Users\ResTIC16\Documents\registros_2025.xlsx")

perfil = pd.read_excel(r"C:\Users\ResTIC16\Documents\perfil_clienteV2.xlsx")


for df in [df23, df24, df25, perfil]:
    df["ID_CLIENTE"] = df["ID_CLIENTE"].astype(str).str.strip()



atendimentos = pd.concat([df23, df24, df25], ignore_index=True)

print("Total de atendimentos:", len(atendimentos))



df_final = atendimentos.merge(
    perfil,
    on="ID_CLIENTE",
    how="inner"   
)

# ===========================
# 5. Exportar planilha final
# ===========================

df_final.to_excel(r"C:\Users\ResTIC16\Documents\dataset3_final.xlsx", index=False)

df_final.head()
```
